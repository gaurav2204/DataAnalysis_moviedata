### Research Question 3 (Top 10 most profitable and non profitable movies inbetween 1960 to 2015)

```python
# 01. Top ten most profitable movie all the years 1960 to 2015
profit_sort = df_movie['profit'].sort_values(ascending=False)[:10]

most_profitable=pd.DataFrame()
titles=[]
profit=[]
for i in profit_sort.index:
    titles.append(df_movie.loc[i,'original_title'])
    profit.append(profit_sort.loc[i])
most_profitable['Movies Titles']=titles
most_profitable['Profit']=profit
most_profitable.set_index('Movies Titles',inplace=True)
print(most_profitable)
```

```python
# 02. Top ten Non-profitable movie all the years 1960 to 2015
loss_sort = df_movie['profit'].sort_values(ascending=True)[:10]

non_profitable=pd.DataFrame()
titles=[]
profit=[]
for i in loss_sort.index:
    titles.append(df_movie.loc[i,'original_title'])
    profit.append(loss_sort.loc[i])
non_profitable['Movies Titles']=titles
non_profitable['Non Profit']=profit
non_profitable.set_index('Movies Titles',inplace=True)
print(non_profitable)
```

```python
# plot show clearly the most profitable top ten movies
most_profitable.plot(kind ='bar',figsize=(8,8),color = 'green')
plt.title('Top 10 highest Profitable movies (1960 - 2015) \n');
plt.ylabel('Profit in billions ($)');
```

```python
# plot show clearly the most profitable top ten movies
non_profitable.plot(kind ='bar',figsize=(8,8), color = 'red')
plt.title('Top 10 Highest Non-Profitable movies (1960 - 2015) \n');
plt.ylabel('Loss in billions ($)');
```

> **observation **:  Both profit and loss clearly shown in data top 10 year in between 1960 to 2015.In graph clearly shown the profit and loss which graph is downwards side is loss show and upwards direction bar graph show profit 

### Research Question 4  (which movie earn maximum and minimum profit in between 1960 to 2015 ?)

```python
# make a function name max_min in which can help out to find the max and min value of column enter value 
def max_min(column):
    #for maximum value
    max_value = df_movie[column].idxmax()
    maximum_value = pd.DataFrame(df_movie.loc[max_value])
    
    #for minimum value
    min_value = df_movie[column].idxmin()
    minimum_value = pd.DataFrame(df_movie.loc[min_value])
    
    #Both max and min value in one place 
    info = pd.concat([maximum_value , minimum_value ], axis=1)
    
    return info

#for profit max and min
max_min('profit')
```

> **observation **: In the above data we find top ten high and non profitable movie all the years 1960 to 2015.
          MOst proitable movie **'Avatar'** with budget 237000000 have 2544505847 profit most protabile movie all time    release year 2009.On the other hand non profitable movie of all time is **'The Warrior's Way'** with budget 425000000 have loss 413912431 release year 2010.
          


### Research Question 5 (Which top 10 movies is Highest in budget and revenue ?  )

```python
# using .nlargest function to get top 10 highest budget 
high_budget=df_movie.nlargest(10, 'budget', keep='first')
high_budget
```

```python
# plot show clearly the most profitable top ten movies
high_budget['budget'].plot(kind ='bar',figsize=(8,8))
plt.title('Top 10 highest budget movies (1960 - 2015) \n');
plt.ylabel('Budget in billions ($)');
```

```python
# using .nlargest function to get top 10 highest revenue 
high_revenue=df_movie.nlargest(10, 'revenue', keep='first')
high_revenue
```

```python
# plot show clearly the most profitable top ten movies
high_revenue['revenue'].plot(kind ='bar',figsize=(8,8))
plt.title('Top 10 highest budget movies (1960 - 2015) \n');
plt.ylabel('Budget in billions ($)');
```
